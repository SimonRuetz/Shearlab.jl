```julia
push!(LOAD_PATH,pwd()*"/../../src") 
import Shearlab
using PyPlot
reload("Shearlab")
```

```julia
# Read Data
n = 512;
# The path of the image
name = "../../data_samples/lena.jpg";
data = load_image(name, n);
#Rescale image in [0,1] summing the 3 arrays in the RGB format
data = Shearlab.rescale(sum(data,3));
```

```julia
# Reduce one dimension 
data = data[:,:,1];
Shearlab.imageplot(data);
```

```julia
sizeX = size(data,1);
sizeY = size(data,2);
```

```julia
rows = sizeX;
cols = sizeY;
X = data; 
nScales = 2;
```

```julia
# Compute the corresponding shearlet system
tic()
shearletSystem = Shearlab.SLgetShearletSystem2D(rows,cols,nScales);
toc()
```

```julia
# Compute the coefficients
tic()
coeffs = Shearlab.SLsheardec2D(X,shearletSystem);
toc()
```

```julia
# Look at the third shsearlet coefficient
Shearlab.imageplot(real(coeffs[:,:,3]))
```

```julia
size(coeffs)
```

```julia
# Look at the last shearlet coefficient
Shearlab.imageplot(real(coeffs[:,:,17]))
```

```julia
# Make the recovery
tic()
Xrec=Shearlab.SLshearrec2D(coeffs,shearletSystem);
toc()
```

```julia
# The recovery is very good
Shearlab.imageplot(Xrec);
```
